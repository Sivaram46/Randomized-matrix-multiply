# Analysis of Random Matrix Multiplication

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from random_LA import random_matmul
```

```python
N_SAMPLES = 5000
m, n, p = 700, 1000, 800

A = np.random.rand(m, n)
B = np.random.rand(n, p)

C_norm, error_norm = random_matmul(
    A, B, n_samples=N_SAMPLES, sampling='norm',
    analysis=True
)
```

```python
plt.plot(np.arange(len(error_norm)), error_norm)
plt.title('Column norm sampling')
plt.xlabel('n_samples (*50)')
plt.ylabel('Error')
plt.show()
```

```python
C_uniform, error_uniform = random_matmul(
    A, B, n_samples=N_SAMPLES, sampling='uniform',
    analysis=True
)
```

```python
plt.plot(np.arange(len(error_uniform)), error_uniform)
plt.title('Uniform column sampling')
plt.xlabel('n_samples (*50)')
plt.ylabel('Error')
plt.show()
```

```python
plt.plot(np.arange(len(error_uniform)), error_uniform, label='uniform')
plt.plot(np.arange(len(error_norm)), error_norm, label='norm')
plt.title('Column norm sampling')
plt.xlabel('n_samples (*50)')
plt.ylabel('Error')
plt.legend()
plt.show()
```

From the above plot, it is evident that the column norm sampling works better than uniform column sampling which is obvious. But, the errors between column norm sampling and uniform sampling woulbe be significant for **real time data**.
